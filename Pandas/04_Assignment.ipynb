```python
merge_data = pd.merge(sales,customer, on= 'CustomerID')

laptop_orders = merge_data [merge_data['Product'] == 'Laptop']

print(laptop_orders)
```

```python
merge_data['Revenue'] = merge_data['Price'] * merge_data['Quantity']
revenue_by_region = merge_data.groupby('Region')['Revenue'].sum()
print(revenue_by_region)
```

```python
avg_quantity = merge_data.groupby('Category')['Quantity'].mean()
print(avg_quantity)
```

```python
customer_revenue = merge_data.groupby('CustomerName')['Revenue'].sum()
top_customer = customer_revenue.idxmax()

print(top_customer)
```
